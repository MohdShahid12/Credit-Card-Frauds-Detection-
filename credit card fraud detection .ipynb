```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting columns excluding 'Time', 'Amount', and 'Class'
cols_to_plot = [col for col in df.columns if col not in ['Time', 'Amount', 'Class']]
df_temp = df[cols_to_plot]

# Create histogram plots
fig, ax = plt.subplots(ncols=4, nrows=7, figsize=(20, 50))
index = 0
ax = ax.flatten()

for col in df_temp.columns:
    sns.histplot(df_temp[col], ax=ax[index], kde=True)  # Use histplot for histograms with kernel density estimate
    index += 1

plt.tight_layout(pad=0.5, w_pad=0.5, h_pad=5)
plt.show()
```
